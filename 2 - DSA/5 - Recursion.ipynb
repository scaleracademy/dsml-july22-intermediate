```python
def factorial(n):
    if n < 0:
        return "Factorial only defined on positive integers!"
    
    if n == 1 or n == 0:
        return 1
    
    result = factorial(n-1)
    
    return n * result
```

```python
factorial(1000)
```

```python
factorial(5)
```

```python
def merge_sort(a):
    # Base condition
    if len(a) <= 1:
        return a
    
    n = len(a)
    
    # Splitting Logic
    left = merge_sort(a[:n//2])
    right = merge_sort(a[n//2:])
    
    print(f"left -> {left}")
    print(f"right -> {right}")
    
    # Merging Logic
    i, j = 0, 0
    result = []
    
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
            
    if i < len(left):
        result += left[i:]
    elif j < len(right):
        result += right[j:]
        
    return result
```

```python
a = [4,5,6,1,2,9,0,7,8,11]
```

```python
merge_sort(a)
```
